```python
import spotipy
from spotipy.oauth2 import SpotifyOAuth

scope = "user-library-read"

sp = spotipy.Spotify(auth_manager=SpotifyOAuth(scope=scope))

results = sp.current_user_saved_tracks()
for idx, item in enumerate(results['items']):
    track = item['track']
    print(idx, track['artists'][0]['name'], " – ", track['name'])
```

```python

from collections import Counter


# 2. Get the "Top 50 - Global" Playlist (This is the official static ID)
playlist_id = '3MAUEOioSuV6ZlWRPmVp2K'
results = sp.playlist_items(playlist_id)

all_tracks = []
genre_list = []

print(f"--- Fetching Global Top 50 ---")

for item in results['items']:
    track = item['track']
    track_name = track['name']
    artist_name = track['artists'][0]['name']
    artist_id = track['artists'][0]['id']
    
    all_tracks.append(f"{artist_name} - {track_name}")
    
    # 3. Fetch artist metadata to get genres
    artist_info = sp.artist(artist_id)
    genre_list.extend(artist_info['genres'])

# 4. Process the Top Genres
genre_counts = Counter(genre_list)
top_50_genres = genre_counts.most_common(50)

print("\n### TOP GENRES IN GLOBAL TOP 50 ###")
for i, (genre, count) in enumerate(top_50_genres, 1):
    print(f"{i}. {genre} (appears {count} times)")

print("\n### TOP 50 SONGS ###")
for i, song in enumerate(all_tracks, 1):
    print(f"{i}. {song}")
```

```python


# 2. Search for the top tracks globally (using a 'top' query or filtering by popularity)
# This bypasses the restricted "Official Playlist" ID
print("--- Fetching Global Hits ---")
results = sp.search(q='year:2024-2026', type='track', limit=50, market='US')

track_ids = []
track_details = []

for item in results['tracks']['items']:
    track_details.append({
        'name': item['name'],
        'artist_name': item['artists'][0]['name'],
        'artist_id': item['artists'][0]['id']
    })
    track_ids.append(item['artists'][0]['id'])

# 3. Batch fetch artists (to avoid 50 separate calls and get genres)
# This is faster and avoids rate limits
artist_ids = [t['artist_id'] for t in track_details]
genres = []

# Spotify allows 50 artists per request
artists_data = sp.artists(artist_ids)['artists']

for artist in artists_data:
    genres.extend(artist['genres'])

# 4. Results
genre_counts = Counter(genres)
top_50_genres = genre_counts.most_common(50)

print("\n### TOP GENRES (From Top 50 Hits) ###")
for i, (genre, count) in enumerate(top_50_genres, 1):
    print(f"{i}. {genre} ({count})")

print("\n### THE 50 SONGS ANALYZED ###")
for i, song in enumerate(track_details, 1):
    print(f"{i}. {song['artist_name']} - {song['name']}")
```
