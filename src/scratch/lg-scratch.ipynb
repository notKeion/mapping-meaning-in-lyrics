# Lyric Genius API Testing Zone
Here we will be testing the Genius API to see if we can pull lyrics and other song information for our project.

```python
from lyricsgenius import Genius

# Replace with your actual Genius Access Token
ACCESS_TOKEN = "PUT IN YOUR ENV"

# Initialize the Genius API
genius = Genius(ACCESS_TOKEN)

# Optional: Remove section headers like [Chorus] or [Verse]
genius.remove_section_headers = True

def get_song_lyrics(title, artist):
    try:
        # Search for the song
        song = genius.search_song(title, artist)
        
        if song:
            print(f"\n--- Lyrics for '{song.title}' by {song.artist} ---\n")
            return song.lyrics
        else:
            return "Song not found."
            
    except Exception as e:
        return f"An error occurred: {e}"

# Usage
lyrics = get_song_lyrics("Bohemian Rhapsody", "Queen")
print(lyrics)
```
